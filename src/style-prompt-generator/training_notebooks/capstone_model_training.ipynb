### Set W&B API Key from Colab Secrets

This cell will retrieve your Weights & Biases API key from Colab's 'Secrets' and set it as an environment variable, making it accessible to `os.environ.get()` within your `train.py` script.

```python
from google.colab import userdata
import os

# Retrieve the API key from Colab secrets
wandb_api_key = userdata.get('WANDB_API_KEY')

if wandb_api_key:
    os.environ['WANDB_API_KEY'] = wandb_api_key
    print("W&B API key loaded from Colab secrets and set as environment variable.")
else:
    print("W&B_API_KEY not found in Colab secrets. Please ensure it's set there.")
```
